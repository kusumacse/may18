```python
import numpy as np
t=np.arange(0.0,2.0,0.1)
v=1+np.sin(2*np.pi*t)
plt.xlabel('time(s)')
plt.ylabel('voltage(mv)')
plt.grid(True)
plt.plot(t,v)
plt.savefig('Figures/time-voltage.png')
plt.show()
```

###plot a line graph on income vs year for the state of california 

```python
import pandas as pd
df = pd.read_csv('Untitled Folder/sample.csv')
y=df.values[4]
x=df.columns
x
y
```

```python
years=[]
for i in range(2,len(x)):
    years.append(x[i])
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
    income
plt.xlabel('year')
plt.ylabel('income')
plt.plot(years,income)
plt.grid(True)
plt.show()
    
```

## Histogram

```python
width=1/2
plt.width=1/2
plt.title('Income of california')
plt.bar(years,income,width,color='Gray')
plt.show()
```

##Generate a scatter plot for 50 random nos


```python
n=50
x=np.random.rand(n)
y=np.random.rand(n)
colorlist = np.random.rand(n)
arealist=np.pi*(15*np.random.rand(20))**2
plt.scatter(x,y, s=arealist,c=colorlist,alpha=1)
plt.savefig('Figures/scatterplot.png')
```
